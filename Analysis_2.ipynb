```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent="aitor.cengon@gmail.com")
```

```python
db_rentamedia_distritos['city_coord']=db_rentamedia_distritos['Unnamed: 0'].apply(geolocator.geocode)
```

```python
db_rentamedia_distritos['latitude']=None
db_rentamedia_distritos['longitude']=None
```

```python
db_rentamedia_distritos=db_rentamedia_distritos.drop([0,1,2,3,4,5,6,7])
```

```python
db_rentamedia_distritos['latitude'] = db_rentamedia_distritos['city_coord'].apply(lambda x: (x.latitude))
db_rentamedia_distritos['longitude'] = db_rentamedia_distritos['city_coord'].apply(lambda x: (x.longitude))
```

```python
db_rentamedia_distritos
```

```python
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 3'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 4'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 5'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 6'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 7'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 8'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 9'],axis=1)
db_rentamedia_distritos=db_rentamedia_distritos.drop(['Unnamed: 10'],axis=1)
```

```python
db_rentamedia_distritos
```

```python
db_rentamedia_distritos = db_rentamedia_distritos.rename(columns={'Unnamed: 0': 'Distritos'})
db_rentamedia_distritos = db_rentamedia_distritos.rename(columns={'Unnamed: 1': 'Renta_media_persona'})
db_rentamedia_distritos = db_rentamedia_distritos.rename(columns={'Unnamed: 2': 'Renta_media_hogar'})
```

```python
db_rentamedia_distritos
```

```python
db_rentamedia_distritos.Distritos=db_rentamedia_distritos.Distritos.str.replace(", Madrid", "")
```

```python
db_rentamedia_distritos
```

```python
db_distritos= pd.read_csv("db_merge_distritos.csv")
db_distritos=db_distritos.drop(['Unnamed: 0'],axis=1)
db_distritos = db_distritos.rename(columns={'district': 'Distritos'})
db_distritos
```

```python
db_datos_distritos=db_distritos.merge(db_rentamedia_distritos,how="left", on=['Distritos'])
db_datos_distritos
```

```python
#That's a very interesting function because it allows us to see how many years of incomes you have to dedicate to pay for a home in each area 
```

```python
def years(x):
    
    x['years_to_buy']=x['buy_price']/(x['Renta_media_persona'])
    
    return x
```

```python
db_datos_distritos= db_datos_distritos.apply(years,axis=1)
```
